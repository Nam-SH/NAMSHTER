# 1699. 제곱수의 합

<br>

![image](https://user-images.githubusercontent.com/50367487/64433568-b4c8bf80-d0f9-11e9-9875-06fbf94a5ab8.png)

```python
n = int(input())
d = [0]*(n+1)
for i in range(1, n+1):
    d[i] = i
    j = 1
    while j*j <= i:
        if d[i] > d[i-j*j]+1:
            d[i] = d[i-j*j]+1
        j += 1
print(d[n])
```

# 2225. 합분해

<br>

![image](https://user-images.githubusercontent.com/50367487/64433580-bc886400-d0f9-11e9-82e8-433bb41f2895.png)

```python
mod = 1000000000
n,k = map(int,input().split())
d = [[0]*(n+1) for _ in range(k+1)]
d[0][0] = 1
for i in range(1, k+1):
    for j in range(0, n+1):
        for l in range(0, j+1):
            d[i][j] += d[i-1][j-l]
        d[i][j] %= mod
print(d[k][n])
```

# 13707. 합분해 2

<br>

![image](https://user-images.githubusercontent.com/50367487/64433590-c1e5ae80-d0f9-11e9-81c9-844f9b8065d6.png)

```python
mod = 1000000000
n,k = map(int,input().split())
d = [[0]*(n+1) for _ in range(k+1)]
d[0][0] = 1
for i in range(1, k+1):
    for j in range(0, n+1):
        d[i][j] = d[i-1][j]
        if j-1 >= 0:
            d[i][j] += d[i][j-1]
        d[i][j] %= mod
print(d[k][n])
```
