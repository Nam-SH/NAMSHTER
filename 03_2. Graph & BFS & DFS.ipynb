# 1260. DFS & BFS

<br>

![image](https://user-images.githubusercontent.com/50367487/63646210-15dfb300-c74a-11e9-9b23-77a9301b4be5.png)

```python
from collections import deque
n, m, start = 4, 5, 1
a = [[], [2, 4], [1, 3], [1, 3], [1, 2]]
check = [False] * (n+1)

def bfs(start):
    check = [False] * (n+1)
    q = deque()
    q.append(start)
    check[start] = True
    while q:
        x = q.popleft()
        print(x, end=' ')
        for y in a[x]:
            if check[y] == False:
                check[y] = True
                q.append(y)

bfs(start)
```

```python
# 인접 리스트 소스

from collections import deque
n,m,start = map(int,input().split())
a = [[] for _ in range(n+1)]
check = [False] * (n+1)

for _ in range(m):
    u,v = map(int,input().split())
    a[u].append(v)
    a[v].append(u)

    for i in range(n):
        a[i].sort()

def dfs(x):
    global check
    check[x] = True
    print(x, end=' ')
    for y in a[x]:
        if check[y] == False:
            dfs(y)

def bfs(start):
    check = [False] * (n+1)
    q = deque()
    q.append(start)
    check[start] = True
    while q:
        x = q.popleft()
        print(x, end=' ')
        for y in a[x]:
            if check[y] == False:
                check[y] = True
                q.append(y)

dfs(start)
print()
bfs(start)
print()
```

```python
# 간선 리스트 소스

from collections import deque
n,m,start = map(int,input().split())
edges = []
check = [False] * (n+1)
for _ in range(m):
    u,v = map(int,input().split())
    edges.append((u,v))
    edges.append((v,u))
m *= 2
edges.sort()
cnt = [0]*(n+1)

for u, v in edges:
    cnt[u] += 1

for i in range(1, n+1):
    cnt[i] += cnt[i-1]

def dfs(x):
    global check
    check[x] = True
    print(x, end=' ')
    for i in range(cnt[x-1], cnt[x]):
        y = edges[i][1]
        if check[y] == False:
            dfs(y)

def bfs(start):
    check = [False] * (n+1)
    q = deque()
    q.append(start)
    check[start] = True
    while q:
        x = q.popleft()
        print(x, end=' ')
        for i in range(cnt[x-1],cnt[x]):
            y = edges[i][1]
            if check[y] == False:
                check[y] = True
                q.append(y)

dfs(start)
print()
bfs(start)
print()
```

```python
# 비재귀 구현 소스

from collections import deque
n,m,start = map(int,input().split())
a = [[] for _ in range(n+1)]
for _ in range(m):
    u,v = map(int,input().split())
    a[u].append(v)
    a[v].append(u)
for i in range(n):
    a[i].sort()

def dfs(node):
    check = [False] * (n+1)
    stack = []
    stack.append((node,0))
    check[node] = True
    print(node, end=' ')
    while stack:
        x,start = stack.pop()
        for i in range(start, len(a[x])):
            y = a[x][i]
            if check[y] == False:
                print(y, end=' ')
                check[y] = True
                stack.append((x,i+1))
                stack.append((y,0))
                break

def bfs(start):
    check = [False] * (n+1)
    q = deque()
    q.append(start)
    check[start] = True
    while q:
        x = q.popleft()
        print(x, end=' ')
        for y in a[x]:
            if check[y] == False:
                check[y] = True
                q.append(y)

dfs(start)
print()
bfs(start)
print()
```

# 11724. 연결 요소의 개수

<br>

![image](https://user-images.githubusercontent.com/50367487/63649875-1b56f080-c77e-11e9-90bc-4b01c2421419.png)

```python
n,m = map(int,input().split())
a = [[] for _ in range(n + 1)]
check = [False] * (n + 1)

for _ in range(m):
    u,v = map(int,input().split())
    a[u].append(v)
    a[v].append(u)

def dfs(x):
    check[x] = True
    for y in a[x]:
        if check[y] == False:
            dfs(y)

ans = 0
for i in range(1, n + 1):
    if not check[i]:
        dfs(i)
        ans += 1
print(ans)
```

# 1707. 이분 그래프

<br>

![image](https://user-images.githubusercontent.com/50367487/63646217-24c66580-c74a-11e9-9ebd-36b311482dc9.png)

```python
for tc in range(int(input())):
    n,m = map(int, input().split())
    a = [[] for _ in range(n + 1)]
    check = [0] * (n + 1)

    for _ in range(m):
        u, v = map(int, input().split())
        a[u].append(v)
        a[v].append(u)

    def dfs(x, c):
        check[x] = c
        for y in a[x]:
            if check[y] == 0 and not dfs(y, 3-c): return False
            elif check[y] == check[x]: return False
        return True

    ans = True
    for i in range(1, n + 1):
        if check[i] == 0 and not dfs(i, 1):
            ans = False
            
    print('#{} {}'.format(tc + 1, 'YES' if ans else 'NO'))
```
