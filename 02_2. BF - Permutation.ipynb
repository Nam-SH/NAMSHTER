# 10972. 다음 순열구하기

<br>

![image](https://user-images.githubusercontent.com/50367487/63637488-87225600-c6b7-11e9-96bb-93f254f0183b.png)

```python
def next_permutation(a):
    i = len(a)-1
    while i > 0 and a[i-1] >= a[i]:
        i -= 1
    if i <= 0:
        return False
    j = len(a)-1
    while a[j] <= a[i-1]:
        j -= 1

    a[i-1],a[j] = a[j],a[i-1]

    j = len(a)-1
    while i < j:
        a[i],a[j] = a[j],a[i]
        i += 1
        j -= 1

    return True

n = int(input())
a = list(map(int,input().split()))
if next_permutation(a): print(' '.join(map(str,a)))
else: print(-1)
```

# 10819. 차이를 최대로

<br>

![image](https://user-images.githubusercontent.com/50367487/63637126-acad6080-c6b3-11e9-9401-2af21acdbfc6.png)

```python
def next_permutation(a):
    i = len(a)-1
    while i > 0 and a[i-1] >= a[i]: 
        i -= 1
    if i <= 0: return False
    
    j = len(a) - 1
    
    while a[j] <= a[i-1]:
        j -= 1

    a[i-1], a[j] = a[j], a[i-1]

    j = len(a) - 1
    
    while i < j:
        a[i], a[j] = a[j], a[i]
        i += 1
        j -= 1

    return True

def calc(a):
    ans = 0
    for i in range(1, len(a)):
        ans += abs(a[i]-a[i-1])
    return ans

n = int(input())
a = list(map(int,input().split()))
a.sort()
ans = 0
while True:
    temp = calc(a)
    ans = max(ans, temp)
    if not next_permutation(a): break
```

# 10971. 외판원 순회

<br>

![image](https://user-images.githubusercontent.com/50367487/63637101-6821c500-c6b3-11e9-8b1b-21164fd060e1.png)

```python
def next_permutation(a):
    i = len(a)-1
    while i > 0 and a[i-1] >= a[i]:
        i -= 1
    if i <= 0:
        return False
    j = len(a)-1
    while a[j] <= a[i-1]:
        j -= 1

    a[i-1],a[j] = a[j],a[i-1]

    j = len(a)-1
    while i < j:
        a[i],a[j] = a[j],a[i]
        i += 1
        j -= 1

    return True

n = int(input())
w = [list(map(int,input().split())) for _ in range(n)]
d = list(range(n))
ans = 2147483647
while True:
    ok = True
    s = 0
    for i in range(0, n-1):
        if w[d[i]][d[i+1]] == 0:
            ok = False
            break
        else:
            s += w[d[i]][d[i+1]]
    if ok and w[d[-1]][d[0]] != 0:
        s += w[d[-1]][d[0]]
        ans = min(ans, s)
    if not next_permutation(d): break
    if d[0] != 0: break
print(ans)
```
