````python
import sys
n = 20
m = int(sys.stdin.readline())
s = 0
for _ in range(m):
    op, *num = sys.stdin.readline().split()
    if len(num) > 0:
        x = int(num[0])-1
    if op == 'add':
        s = (s | (1 << x))
    elif op == 'remove':
        s = (s & ~(1 << x))
    elif op == 'check':
        res = (s & (1 << x))
        if res > 0:
            sys.stdout.write('1\n')
        else:
            sys.stdout.write('0\n')
    elif op == 'toggle':
        s = (s ^ (1 << x))
    elif op == 'all':
        s = (1 << n) - 1
    else:
        s = 0
````

# 1182. 부분수열의 합

<br>

![image](https://user-images.githubusercontent.com/50367487/63637058-cc905480-c6b2-11e9-84ac-6d086581f4fb.png)

````python
n,m = map(int,input().split())
a = list(map(int,input().split()))
ans = 0
for i in range(1, (1<<n)):
    s = sum(a[k] for k in range(n) if (i & (1<<k)) > 0)
    if m == s:
        ans += 1
print(ans)
````
