# 2609. 최대공약수와 최소공배수

<br>

![image](https://user-images.githubusercontent.com/50367487/63644008-46f3bf80-c719-11e9-9354-290f6926f161.png)

```python
def gcd(x, y):
    if y == 0:
        return x
    else:
        return gcd(y, x%y)
a,b = map(int, input().split())
g = gcd(a, b)
print(g)
print(a*b//g)
```

# 1934. 최소공배수

<br>

![image](https://user-images.githubusercontent.com/50367487/63644010-5a9f2600-c719-11e9-9f48-6ded0e0aeac0.png)

```python
def gcd(x, y):
    if y == 0:
        return x
    else:
        return gcd(y, x%y)
t = int(input())
for _ in range(t):
    a,b = map(int, input().split())
    g = gcd(a, b)
    print(a*b//g)
```

# 9613. GCD 합

<br>

![image](https://user-images.githubusercontent.com/50367487/63644011-625eca80-c719-11e9-968e-2544e847d0bd.png)

```python
def gcd(x, y):
    if y == 0:
        return x
    else:
        return gcd(y, x%y)
t = int(input())
for _ in range(t):
    a = list(map(int, input().split()))
    n = a[0]
    a = a[1:]
    ans = 0
    for i in range(0, n-1):
        for j in range(i+1, n):
            ans += gcd(a[i], a[j])
    print(ans)
```

# 1978. 소수 찾기

<br>

![image](https://user-images.githubusercontent.com/50367487/63644012-67bc1500-c719-11e9-8fdf-e04436f74b92.png)

```python
def is_prime(x):
    if x < 2:
        return False
    i = 2
    while i*i <= x:
        if x % i == 0:
            return False
        i += 1
    return True

n = int(input())
a = list(map(int,input().split()))
ans = 0
for x in a:
    if is_prime(x):
        ans += 1
print(ans)
```

# 1929. 소수 구하기

<br>

![image](https://user-images.githubusercontent.com/50367487/63644013-6c80c900-c719-11e9-824d-0f854408e316.png)

```python
MAX = 1000000
check = [0]*(MAX+1)
check[0] = check[1] = True

for i in range(2, MAX+1):
    if not check[i]:
        j = i+i
        while j <= MAX:
            check[j] = True
            j += i
m, n = map(int,input().split())
for i in range(m, n+1):
    if check[i] == False:
        print(i)
```

# 6588. 골드바흐의 추측

<br>

![image](https://user-images.githubusercontent.com/50367487/63644014-75719a80-c719-11e9-9ae7-86c2d8476ff7.png)

```python
MAX = 1000000
check = [0]*(MAX+1)
check[0] = check[1] = True
prime = []
for i in range(2, MAX+1):
    if not check[i]:
        prime.append(i)
        j = i+i
        while j <= MAX:
            check[j] = True
            j += i
prime = prime[1:]
while True:
    n = int(input())
    if n == 0:
        break
    for p in prime:
        if check[n-p] == False:
            print("{0} = {1} + {2}".format(n, p, n-p))
            break
```
