```python
# bottom-up 방식

n = int(input())
d = [0]*(n+1)
d[1] = 0
for i in range(2, n+1):
    d[i] = d[i-1] + 1
    if i%2 == 0 and d[i] > d[i//2] + 1:
        d[i] = d[i//2] + 1
    if i%3 == 0 and d[i] > d[i//3] + 1:
        d[i] = d[i//3] + 1
print(d[n])
```

# 11726. 2xn 타일링

<br>

![image](https://user-images.githubusercontent.com/50367487/64432795-31f33500-d0f8-11e9-9666-343e852d6700.png)

```python
n = int(input())
d = [0]*1001
d[0] = 1
d[1] = 1
for i in range(2, n+1):
    d[i] = d[i-1] + d[i-2]
    d[i] %= 10007
print(d[n])
```

# 9095. 1, 2, 3 더하기

<br>

![image](https://user-images.githubusercontent.com/50367487/64432815-39b2d980-d0f8-11e9-8679-09df1d5aa264.png)

```python
d = [0]*11
d[0] = 1
for i in range(1, 11):
    if i-1 >= 0:
        d[i] += d[i-1]
    if i-2 >= 0:
        d[i] += d[i-2]
    if i-3 >= 0:
        d[i] += d[i-3]

t = int(input())
for _ in range(t):
    n = int(input())
    print(d[n])
```

# 15988. 1, 2, 3 더하기 3

<br>

![image](https://user-images.githubusercontent.com/50367487/64432833-41727e00-d0f8-11e9-9121-a552b2585189.png)

```python
d = [0]*1000001
mod = 1000000009
d[0] = 1
for i in range(1, 1000000+1):
    if i-1 >= 0:
        d[i] += d[i-1]
    if i-2 >= 0:
        d[i] += d[i-2]
    if i-3 >= 0:
        d[i] += d[i-3]
    d[i] %= mod

t = int(input())
for _ in range(t):
    n = int(input())
    print(d[n])
```
