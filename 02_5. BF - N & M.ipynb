```python
n,m = map(int,input().split())
c = [False]*(n+1)
a = [0]*m

def go(index, start, n, m):
    if index == m:
        print(*a)
        return
    for i in range(start, n+1):
        if c[i]:
            continue
        c[i] = True
        a[index] = i
        go(index+1, i+1, n, m)
        c[i] = False

go(0,1,n,m)
```

```python
def find(idx, select, n, m):
    if idx == m:
        print(*a)
        return
    if select > n:
        return
    
    a[idx] = select
    find(idx + 1, select + 1, n, m)
    
    a[idx] = 0 
    find(idx, select + 1, n, m)

n, m = map(int, input().split())
a = [0] * m

find(0, 1, n, m)
```

# 15663. N과 M (9)

<br>

![image](https://user-images.githubusercontent.com/50367487/63644291-aa352000-c720-11e9-9bef-4e8d2890acc6.png)

```python
import sys
n,m = map(int,input().split())
num = list(map(int,input().split()))
num.sort()
c = [False]*n
a = [0]*m
d = []

def go(index, n, m):
    if index == m:
        d.append(tuple(a))
        return
    for i in range(n):
        if c[i]: continue
        c[i] = True
        a[index] = num[i]
        go(index+1, n, m)
        c[i] = False
go(0,n,m)
d = sorted(list(set(d)))
for v in d:
    print(' '.join(map(str,v)), end = ' ')
    print()
```

# 15664. N과 M (10)

<br>

![image](https://user-images.githubusercontent.com/50367487/63644292-b0c39780-c720-11e9-869a-0ca265afd070.png)

```python
from collections import Counter

def go(index, start, n, m):
    if index == m:
        print(' '.join(map(str,a)))
        return
    for i in range(start,n):
        if cnt[i] > 0:
            cnt[i] -= 1
            a[index] = num[i]
            go(index + 1, i, n, m)
            cnt[i] += 1

n,m = map(int,input().split())
temp = list(map(int,input().split()))
temp = list(Counter(temp).items())
temp.sort()

n = len(temp)
num, cnt = map(list,zip(*temp))
a = [0] * m

go(0,0,n,m)
```
