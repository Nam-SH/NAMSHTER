# 2309. 일곱 난쟁이

<br>

![image](https://user-images.githubusercontent.com/50367487/63636707-b9c75100-c6ad-11e9-8405-7723186beada.png)

```python
import sys; sys.stdin = open('2309.txt', 'r')

arr = [int(input()) for _ in range(9)]
arr.sort()
for i in range(9):
    for j in range(i + 1, 9):
        if (sum(arr) - arr[i] - arr[j]) == 100:
            for k in range(9):
                if k == i or k == j:
                    continue
                print(arr[k])
            sys.exit(0)
```

# 1476. 날짜 계산

<br>

![image](https://user-images.githubusercontent.com/50367487/63636712-d3689880-c6ad-11e9-8e18-05cfb1e695d9.png)

```python
import sys; sys.stdin = open('1476.txt', 'r')

E, S, M = map(int, input().split())
e, s, m = 1, 1, 1
count = 1

while 1:

    if e == E and s == S and m == M:
        print(count)
        break

    e += 1; s += 1; m += 1
    if e == 16: e = 1
    if s == 29: s = 1
    if m == 20: m = 1
    count += 1
```
