# 11502. 카드 구매하기

<br>

![image](https://user-images.githubusercontent.com/50367487/64432927-71ba1c80-d0f8-11e9-9db6-4a822ed3f01e.png)

```python
n = int(input())
a = [0] + list(map(int,input().split()))
d = [0]*(n+1)
for i in range(1, n+1):
    for j in range(1, i+1):
        d[i] = max(d[i],d[i-j]+a[j])
print(d[n])
```

# 16194. 카드 구매하기 2

<br>

![image](https://user-images.githubusercontent.com/50367487/64433178-e8efb080-d0f8-11e9-905e-ab447512791f.png)

```python
n = int(input())
a = [0] + list(map(int,input().split()))
d = [-1]*(n+1)
d[0] = 0
for i in range(1, n+1):
    for j in range(1, i+1):
        if d[i] == -1 or d[i] > d[i-j]+a[j]:
            d[i] = d[i-j]+a[j]
print(d[n])
```

# 15990. 1, 2, 3 더하기 5

<br>

![image](https://user-images.githubusercontent.com/50367487/64433193-ee4cfb00-d0f8-11e9-9ffb-16ee8b17117a.png)

```python
limit = 100000
d = [[0]*4 for _ in range(limit+1)]
mod = 1000000009
for i in range(1, limit+1):
    if i-1 >= 0:
        d[i][1] = d[i-1][2] + d[i-1][3]
        if i == 1:
            d[i][1] = 1
    if i-2 >= 0:
        d[i][2] = d[i-2][1] + d[i-2][3]
        if i == 2:
            d[i][2] = 1
    if i-3 >= 0:
        d[i][3] = d[i-3][1] + d[i-3][2]
        if i == 3:
            d[i][3] = 1
    d[i][1] %= mod
    d[i][2] %= mod
    d[i][3] %= mod

t = int(input())
for _ in range(t):
    n = int(input())
    print(sum(d[n])%mod)
```

# 10844. 쉬운 계단 수

<br>

![image](https://user-images.githubusercontent.com/50367487/64433198-f4db7280-d0f8-11e9-8733-7475478d0c37.png)

```python
d = [[0]*10 for _ in range(100+1)]
mod = 1000000000 
n = int(input())
for i in range(1, 10):
    d[1][i] = 1
for i in range(2, n+1):
    for j in range(10):
        d[i][j] = 0
        if j-1 >= 0:
            d[i][j] += d[i-1][j-1]
        if j+1 <= 9:
            d[i][j] += d[i-1][j+1]
        d[i][j] %= mod
ans = sum(d[n]) % mod
print(ans)
```

# 11057. 오르막 수

<br>

![image](https://user-images.githubusercontent.com/50367487/64433222-fad15380-d0f8-11e9-9d0c-20bf18c6ee5f.png)

```python
d = [[0]*10 for _ in range(1001)]
mod = 10007
n = int(input())
for i in range(10):
    d[1][i] = 1
for i in range(2, n+1):
    for j in range(10):
        for k in range(j+1):
            d[i][j] += d[i-1][k]
            d[i][j] %= mod
ans = sum(d[n])
print(ans%mod)
```
