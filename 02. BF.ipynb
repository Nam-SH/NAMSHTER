# 9095. 1, 2, 3 더하기

<br>
    
![image](https://user-images.githubusercontent.com/50367487/63632859-b5814080-c679-11e9-9ae7-bc011c409782.png)

```python
def go(s, goal):
    if s > goal: return 0
    if s == goal: return 1
    
    now = 0
    for i in range(1, 4):
        now += go(s+i, goal)
    return now

t = int(input())
for _ in range(t):
    n = int(input())
    print(go(0, n))
```

# 1759. 암호만들기

<br>

![image](https://user-images.githubusercontent.com/50367487/63632972-b4e9a980-c67b-11e9-8634-ab042d37399a.png)

```python
def check(password):
    ja = mo = 0
    for x in password:
        if x in 'aeiou': mo += 1
        else: ja += 1
    return ja >= 2 and mo >= 1

def go(n, alpha, password, i):
    if len(password) == n:
        if check(password): print(password)
        return
    if i == len(alpha): return
    go(n, alpha, password + alpha[i], i+1)
    go(n, alpha, password, i+1)

n, m = map(int,input().split())
data = input().split()
data.sort()
go(n, data, "", 0)
```

#  6603. 로또

<br>

![image](https://user-images.githubusercontent.com/50367487/63633463-e23a5580-c683-11e9-8348-49f955669711.png)

```python
def solve(data, index, lotto):
    if len(lotto) == 6:
        print(' '.join(map(str,lotto)))
        return
    if index == len(data):
        return
    solve(data, index+1, lotto+[data[index]])
    solve(data, index+1, lotto)

    
while True:
    n, *data = map(int,input().split())
    if n == 0:
        break
    solve(data, 0, [])
    print()
```

# 1182. 부분수열의 합

<br>

![image](https://user-images.githubusercontent.com/50367487/63634293-52020d80-c68f-11e9-91d9-bf096b9f7720.png)

```python
n, m = map(int,input().split())
data = list(map(int,input().split()))
ans = 0

def go(i, s):
    global ans
        
    if i == n:
        if s == m:
            ans += 1
        return
    go(i + 1, s + data[i])
    go(i + 1, s)

go(0, 0)
if m == 0: ans -= 1 # 공집합도 포함되므로;; 크기가 양수라는 조건이 달려있음;;
print(ans)
```

# 14501. 퇴사

<br>

![image](https://user-images.githubusercontent.com/50367487/63634310-88d82380-c68f-11e9-92da-844c48cea0b2.png)

```python
# inf = -10**9
n = int(input())
t = [0]*(n+1)
p = [0]*(n+1)
for i in range(1, n+1):
    t[i], p[i] = map(int,input().split())
ans = 0

def go(day, sal):
    global ans
    if day == n+1:
        ans = max(ans, sal)
        return
    if day > n+1:
        return
    go(day+1, sal)
    go(day+t[day], sal+p[day])

go(1, 0)
print(ans)
```

# 15658. 연산자 끼워 넣기 2

<br>

![image](https://user-images.githubusercontent.com/50367487/63634355-1f0c4980-c690-11e9-84fd-564366c321d8.png)

```python
def calc(a, index, cur, plus, minus, mul, div):
    if index == len(a):
        return (cur, cur)
    res = []
    if plus > 0:
        res.append(calc(a,index+1,cur+a[index],plus-1,minus,mul,div))
    if minus > 0:
        res.append(calc(a,index+1,cur-a[index],plus,minus-1,mul,div))
    if mul > 0:
        res.append(calc(a,index+1,cur*a[index],plus,minus,mul-1,div))
    if div > 0:
        if cur >= 0:
            res.append(calc(a,index+1,cur//a[index],plus,minus,mul,div-1))
        else:
            res.append(calc(a,index+1,-(-cur//a[index]),plus,minus,mul,div-1))
    ans = (
        max([t[0] for t in res]),
        min([t[1] for t in res])
    )

    return ans

n = int(input())
a = list(map(int,input().split()))
plus,minus,mul,div = map(int,input().split())
ans = calc(a, 1, a[0], plus, minus, mul, div)
print(ans[0])
print(ans[1])
```
