# 11055. 가장 큰 증가 부분 수열

<br>

![image](https://user-images.githubusercontent.com/50367487/64433376-4f74ce80-d0f9-11e9-8d4e-f6cfe4601bc5.png)

```python
n = int(input())
a = list(map(int,input().split()))
d = [0]*n
for i in range(n):
    d[i] = a[i]
    for j in range(i):
        if a[j] < a[i] and d[j]+a[i] > d[i]:
            d[i] = d[j]+a[i]
print(max(d))
```

# 11722. 가장 긴 감소하는 부분 수열

<br>

![image](https://user-images.githubusercontent.com/50367487/64433390-556aaf80-d0f9-11e9-823a-8a82096390fb.png)

```python
n = int(input())
a = list(map(int,input().split()))
d = [0]*n
for i in range(n):
    d[i] = 1
    for j in range(i):
        if a[j] > a[i] and d[i] < d[j]+1:
            d[i] = d[j]+1
print(max(d))
```

# 11054. 가장 긴 바이토닉 부분 수열

<br>

![image](https://user-images.githubusercontent.com/50367487/64433702-f5283d80-d0f9-11e9-9863-30dd61374efb.png)

```python
n = int(input())
a = list(map(int,input().split()))
d1 = [0]*n
d2 = [0]*n
for i in range(n):
    d1[i] = 1
    for j in range(i):
        if a[j] < a[i] and d1[j]+1 > d1[i]:
            d1[i] = d1[j]+1
for i in range(n-1, -1, -1):
    d2[i] = 1
    for j in range(i+1, n):
        if a[i] > a[j] and d2[j]+1 > d2[i]:
            d2[i] = d2[j]+1
d = [d1[i]+d2[i]-1 for i in range(n)]
print(max(d))
```

# 1912. 연속합

<br>

![image](https://user-images.githubusercontent.com/50367487/64433408-5ac7fa00-d0f9-11e9-8d13-45aa479d87c5.png)

```python
n = int(input())
a = list(map(int,input().split()))
d = [0]*n
for i in range(n):
    d[i] = a[i]
    if i == 0:
        continue
    if d[i] < d[i-1] + a[i]:
        d[i] = d[i-1] + a[i]
print(max(d))
```

# 13398. 연속합 2

<br>


![image](https://user-images.githubusercontent.com/50367487/64433425-5f8cae00-d0f9-11e9-8286-a55691141003.png)

```python
n = int(input())
a = list(map(int,input().split()))
d = [0]*n
dr = [0]*n
for i in range(n):
    d[i] = a[i]
    if i == 0:
        continue
    if d[i] < d[i-1] + a[i]:
        d[i] = d[i-1] + a[i]
for i in range(n-1, -1, -1):
    dr[i] = a[i]
    if i == n-1:
        continue
    if dr[i] < dr[i+1] + a[i]:
        dr[i] = dr[i+1] + a[i]
ans = max(d)
for i in range(1, n-1):
    if ans < d[i-1] + dr[i+1]:
        ans = d[i-1] + dr[i+1]
print(ans)
```
