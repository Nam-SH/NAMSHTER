# 9465. 스티커

<br>

![image](https://user-images.githubusercontent.com/50367487/64433250-0de42380-d0f9-11e9-834b-9d77bdc83c4d.png)

```python
t = int(input())
for _ in range(t):
    n = int(input())
    t1 = [0]+list(map(int,input().split()))
    t2 = [0]+list(map(int,input().split()))
    a = list(zip(t1,t2))
    d = [[0]*3 for _ in range(n+1)]
    for i in range(1, n+1):
        d[i][0] = max(d[i-1])
        d[i][1] = max(d[i-1][0],d[i-1][2])+a[i][0]
        d[i][2] = max(d[i-1][0],d[i-1][1])+a[i][1]
    ans = max(d[n])
    print(ans)
```

# 2156. 포도주 시식

<br>

![image](https://user-images.githubusercontent.com/50367487/64433257-12a8d780-d0f9-11e9-8c2a-9a22246ba91b.png)

```python
n = int(input())
a = [0] + [int(input()) for _ in range(n)]
d = [0] * (n+1)
d[1] = a[1]
if n >= 2:
    d[2] = a[1]+a[2]
for i in range(3,n+1):
    d[i] = d[i-1]
    d[i] = max(d[i], d[i-2]+a[i])
    d[i] = max(d[i], d[i-3]+a[i]+a[i-1])
print(d[n])
```

# 11053. 가장 긴 증가하는 부분 수열

<br>

![image](https://user-images.githubusercontent.com/50367487/64433270-189eb880-d0f9-11e9-8ae8-3dc8ec2466f2.png)

```python
n = int(input())
a = list(map(int,input().split()))
d = [0]*n
for i in range(n):
    d[i] = 1
    for j in range(i):
        if a[j] < a[i] and d[j]+1 > d[i]:
            d[i] = d[j]+1
print(max(d))
```

# 14002. 가장 긴 증가하는 부분 수열 4

<br>

![image](https://user-images.githubusercontent.com/50367487/64433295-22282080-d0f9-11e9-862b-55ebb55e11c0.png)

```python
n = int(input())
a = list(map(int,input().split()))
d = [0]*n
v = [-1]*n
for i in range(n):
    d[i] = 1
    for j in range(i):
        if a[j] < a[i] and d[j]+1 > d[i]:
            d[i] = d[j]+1
            v[i] = j
ans = max(d)
p = [i for i,x in enumerate(d) if x == ans][0]
print(ans)
def go(p):
    if p == -1:
        return
    go(v[p])
    print(a[p],end=' ')
go(p)
print()
```
